## The below code is for training the wounded model with wounded examples. The wounded parts of the signals are stored as wounded_signal_file_{file number}.csv files

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
i = 37
x = 4360
y = 500
data = pd.read_csv(f'./programming/smoothed_signal_file_{i}.csv')
plt.plot(data['Smoothed Voltage'])
plt.grid()
plt.figure(2)
wounded_df = data['Smoothed Voltage'][x:x+y]
plt.plot(data['Smoothed Voltage'][x:x+y])
plt.grid()
```

```python
wounded_df = pd.DataFrame({'Wounded Voltage': wounded_df})
wounded_df.to_csv(f'./wounded/wounded_signal_file_{i}.csv', index=False)
```
