## The below code is for training the wounded model with healthy examples. The healthy parts of the signals are stored as healthy_signal_file_{file number}.csv files

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
i = 35
x = 150000
y = 500

data = pd.read_csv(f'./programming/smoothed_signal_file_{i}.csv')
plt.plot(data['Smoothed Voltage'])
plt.grid()
plt.figure(2)
healthy_df = data['Smoothed Voltage'][x:x+y]
plt.plot(data['Smoothed Voltage'][x:x+y])
plt.grid()
```

```python
healthy_df = pd.DataFrame({'Healthy Voltage': healthy_df})
healthy_df.to_csv(f'./healthy/healthy_signal_file_68.csv', index=False)
```
