# The evaluation of model-2 (variation model) without scaling

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
XT = pd.DataFrame()
subset_list = []
```

```python
i = 29
j = 0
x = 0
y = 10000
m = 4000
start = 200000
end = 500000
gap = 500
data = pd.read_csv(f'./programming/smoothed_signal_file_{i}.csv')
plt.plot(data['Smoothed Voltage'])
plt.grid()
plt.figure(2)
plt.plot(data['Smoothed Voltage'][x:x + y])
plt.grid()
```

```python
x_train = pd.DataFrame(columns=['start_m', 'max_min', 'maxi', 'mini', 'final_initial', 'final', 'initial', 'variance', 'IQR'])

# subset_list = []
num = (end - start)/gap
while (start != end) and (j != num):
    subset = abs(data['Smoothed Voltage'][start:start + 5000])
    subset_list.append(subset)
    start_m = start - m
    #skewness = subset.skew()
    max_min = subset.max() - subset.min()
    maxi = subset.max()
    mini = subset.min()
    final_initial = subset.iloc[4999] - subset.iloc[0]
    final = subset.iloc[4999]
    initial = subset.iloc[0]
    variance = subset.var()
    Q3 = np.quantile(subset, 0.75)
    Q1 = np.quantile(subset, 0.25)
    IQR = Q3 - Q1
    
    x_train.loc[j] = [start_m, max_min, maxi, mini, final_initial, final, initial, variance, IQR]
#     print(start)
    start += gap
    j += 1

print(len(subset_list))
#print(subset_list)
```

```python
XT = pd.concat([XT, x_train], ignore_index=True)
print(XT)
```

```python
XT.to_csv(f'./xtrain/X_Train_variation.csv', index=False)
```

```python
YT = list(np.ones(6681)) + list(np.zeros(5410))
print(YT)
print(len(YT))
```

```python
X_train, X_test, y_train, y_test = train_test_split(XT, YT, test_size=0.3, random_state=42)

svm_classifier = SVC(kernel='rbf') 

svm_classifier.fit(X_train, y_train)

y_pred = svm_classifier.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy:.4f}")
print(X_test)
print(y_test)
print(list(y_pred))
```
